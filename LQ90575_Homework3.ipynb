In the end, the higher epsilon values had a higher average after many time intervals.

# 5.a. Average Observed Reward

### Interpretation
At the beginning, the epsilon greedy method comes out on top over the UCB method. However eventually, the UCB method gets a higher average observed reward because the simulation is doing a more efficient job of deciding which arm to choose. Rather than relying on chance on whether to explore or exploit, the UCB method takes into account the other arms that might still have a chance of actually being the optimal arm. The more often the UCB method picks the arm, the more sure it is of it's actual mean reward. In other words, the UCB method takes into consideration the uncertainty of other arms. As a result, the UCB method learns much better than the epsilon greedy method.

```python
import importlib
import functions as functions
from functions import *
importlib.reload(functions)



solution1 = UCB(initial_value=0, degree_of_seperation=2)
solution2 = epsilon_greedy(epsilon=0.1)
rand = epsilon_greedy(1) # Random selects an arm every time
# Plot graphs here
x = arange(1, len(solution1[0]) + 1) # Needed for the x-axis
plt.title("5.a. Average Observed Reward")
plt.xlabel("Time (t)")
plt.xticks(arange(1, len(solution1[0]), step = len(solution1[0]) / 10))
# plt.yticks
plt.xlim(right = len(solution1[0]))

# Plotting Average Observed Reward here
plt.ylabel("Average Observed Reward Over Time")
plt.plot(x, solution1[0])
plt.plot(x, solution2[0])
plt.plot(x, rand[0])

plt.rc("legend", fontsize = 7)
plt.legend(["UCB Method", "ε Greedy Method", "Random Selection"], loc = "upper left")

print("EXITED WITHOUT ERROR")
```
